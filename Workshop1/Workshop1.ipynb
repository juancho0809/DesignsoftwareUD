#Vehicles

First we make the classes that helps ud to maintain the code more organize

```python
class Engine:
    def __init__(self, name: str, type_of_engine: str, potency: int, weight: int):
        self.name = name
        self.type_of_engine = type_of_engine
        self.potency = potency
        self.weight = weight

class Vehicle:
    def __init__(self, name: str, type_of_engine: str, potency: int, weight: int, chasis: str, model: str, gas_consumption: int, year: int):
        self.engine = Engine(name, type_of_engine, potency, weight)
        self.chasis = chasis
        self.model = model
        self.gas_consumption = gas_consumption
        self.year = year

class Car(Vehicle):
    def __init__(self, name: str, type_of_engine: str, potency: float, weight: float, chasis: str, model: str, gas_consumption: float, year: int, color: str, wheels: int):
        super().__init__(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year)
        self.color = color
        self.wheels = wheels

class Truck(Vehicle):
    def __init__(self, name: str, type_of_engine: str, potency: float, weight: float, chasis: str, model: str, gas_consumption: float, year: int, color: str, wheels: int):
        super().__init__(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year)
        self.color = color
        self.wheels = wheels

class Yacht(Vehicle):
    def __init__(self, name: str, type_of_engine: str, potency: float, weight: float, chasis: str, model: str, gas_consumption: float, year: int, color: str, wheels: int):
        super().__init__(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year)
        self.color = color
        self.wheels = wheels

class Motorcycle(Vehicle):
    def __init__(self, name: str, type_of_engine: str, potency: float, weight: float, chasis: str, model: str, gas_consumption: float, year: int, color: str, wheels: int):
        super().__init__(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year)
        self.color = color
        self.wheels = wheels

Menu = """

Option 1. Create car
Option 2. Create Truck
Option 3. Create Yacht
Option 4. Create Motorcycle
-------------------------------
Option 8. Exit

"""

"""
Select the vehicle who you want to meausure the gas consume
In the calcule gas_consumption, we use three variables, potency, weight and chassis, where
the chassis depend from a conditional
"""


print(Menu)
option = int(input("Press an option: "))
while option != 8:
    if option == 1:
        print("Car")
        name = input("Name: ")
        type_of_engine = input("Type of motor: ")
        potency = float(input("Potency: "))
        weight = float(input("Weight: "))
        chasis = input("Chasis: ")
        model = input("Model: ")
        gas_consumption = 1.1 * potency + 0.2 * weight
        year = int(input("Year: "))
        color = input("Color: ")
        wheels = int(input("Wheels: "))
        car_ = Car(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year, color, wheels)
        if car_.chasis == 'A':
            car_.gas_consumption = car_.gas_consumption - 0.3
        elif car_.chasis == 'B':
            car_.gas_consumption = car_.gas_consumption - 0.5
        print("Vehicle created successfully:", car_.engine.name)
        print("Gas consumption:", car_.gas_consumption)

    elif option == 2:
        print("Truck")
        name = input("Name: ")
        type_of_engine = input("Type of motor: ")
        potency = float(input("Potency: "))
        weight = float(input("Weight: "))
        chasis = input("Chasis: ")
        model = input("Model: ")
        gas_consumption = 1.1 * potency + 0.2 * weight
        year = int(input("Year: "))
        color = input("Color: ")
        wheels = int(input("Wheels: "))
        truck_ = Truck(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year, color, wheels)
        if truck_.chasis == 'A':
            truck_.gas_consumption = truck_.gas_consumption - 0.3
        elif truck_.chasis == 'B':
            truck_.gas_consumption = truck_.gas_consumption - 0.5
        print("Vehicle created successfully:", truck_.engine.name)
        print("Gas consumption:", truck_.gas_consumption)

    elif option == 3:
        print("Yacht")
        name = input("Name: ")
        type_of_engine = input("Type of motor: ")
        potency = float(input("Potency: "))
        weight = float(input("Weight: "))
        chasis = input("Chasis: ")
        model = input("Model: ")
        gas_consumption = 1.1 * potency + 0.2 * weight
        year = int(input("Year: "))
        color = input("Color: ")
        wheels = int(input("Wheels: "))
        yacht_ = Yacht(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year, color, wheels)
        if yacht_.chasis == 'A':
            yacht_.gas_consumption = yacht_.gas_consumption - 0.3
        elif yacht_.chasis == 'B':
            yacht_.gas_consumption = yacht_.gas_consumption - 0.5
        print("Vehicle created successfully:", yacht_.engine.name)
        print("Gas consumption:", yacht_.gas_consumption)

    elif option == 4:
        print("Motorcycle")
        name = input("Name: ")
        type_of_engine = input("Type of motor: ")
        potency = float(input("Potency: "))
        weight = float(input("Weight: "))
        chasis = input("Chasis: ")
        model = input("Model: ")
        gas_consumption = 1.1 * potency + 0.2 * weight
        year = int(input("Year: "))
        color = input("Color: ")
        wheels = int(input("Wheels: "))
        motorcycle_ = Motorcycle(name, type_of_engine, potency, weight, chasis, model, gas_consumption, year, color, wheels)
        if motorcycle_.chasis == 'A':
            motorcycle_.gas_consumption = motorcycle_.gas_consumption - 0.3
        elif motorcycle_.chasis == 'B':
            motorcycle_.gas_consumption = motorcycle_.gas_consumption - 0.5
        print("Vehicle created successfully:", motorcycle_.engine.name)
        print("Gas consumption:", motorcycle_.gas_consumption)

    else:
        print("Option Invalid")

    print(Menu)
    option = int(input("Press an option: "))
```
